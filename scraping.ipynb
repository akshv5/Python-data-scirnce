**üîç What is Web Scraping?**

Web scraping is the process of automatically extracting information from websites. It can be used for things like:

- Collecting product prices

- Gathering data from job sites

- Monitoring stock market prices

- Fetching news headlines, etc.

```python
from dputils.scrape import Scraper , Tag   # pip install dputils
from pprint import pp  # pretty print
```

**- Scraper is a function which is used to make the scraping process**

**- Tag is a class which used to deccide  what to scrape from the page**

**‚úÖ‚úÖExample of the web Scraping using Flipkart!**

```python
url  = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

**Creating object for the web scraping!**

```python
# Create a Scraper object
scraper  = Scraper(url)  

# Get the page content
scraper.get_repeating_date_from_page 
```

```python
# Get the page content

result = scraper.get_repeating_date_from_page(
    target = Tag('div', cls = 'DOjaWF gdgoEp'),
    item = Tag('div', cls = '_75nlfW'),
    title = Tag('div', cls = 'KzDlHZ'),
    price = Tag('div', cls = 'Nx9bqj _4b5DiR')
)  
```

```python
# Displaying list of dicts

result    # Print the result
```

```python
import pandas as pd 
df = pd.DataFrame(result)   
df
```
