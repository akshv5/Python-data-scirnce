```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('country_wise_latest.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataset is loaded into a DataFrame called `df`
top_countries = df.nlargest(10, 'Confirmed')
sns.barplot(data=top_countries, x='Confirmed', y='Country/Region', palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.show()
```

```python
sns.scatterplot(data=df, x='Recovered', y='Deaths', hue='WHO Region', palette='tab10')
plt.title('Deaths vs Recovered by WHO Region')
plt.show()
```

```python
top_5_countries = df.nlargest(5, '1 week % increase')
sns.lineplot(data=top_5_countries, x='Country/Region', y='1 week % increase', marker='o')
plt.title('1 Week % Increase for Top 5 Countries')
plt.show()
```

```python
sns.boxplot(data=df, x='WHO Region', y='Active', palette='Set2')
plt.title('Distribution of Active Cases by WHO Region')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data=df, x='New cases', bins=20, kde=True, color='blue')
plt.title('Distribution of New Cases')
plt.show()
```

```python
df_grouped = df.groupby('WHO Region')['Deaths'].sum().reset_index()
plt.pie(df_grouped['Deaths'], labels=df_grouped['WHO Region'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Deaths by WHO Region')
plt.show()
```

```python
sns.violinplot(data=df, x='WHO Region', y='Deaths / 100 Cases', palette='muted')
plt.title('Deaths / 100 Cases by WHO Region')
plt.xticks(rotation=45)
plt.show()
```

```python
top_new_deaths = df.nlargest(10, 'New deaths')
sns.barplot(data=top_new_deaths, x='New deaths', y='Country/Region', palette='magma')
plt.title('Top 10 Countries by New Deaths')
plt.show()
```
